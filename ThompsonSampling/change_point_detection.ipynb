```python
import ruptures as rpt
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = np.random.normal(5, 1, size=20)
data2 = np.random.normal(10, 1, size=30)
data = np.concatenate((data1, data2))
```

```python
expected_mean = 5
for i in range(len(data) - 10):
    # algo = rpt.Pelt(model="normal").fit(data[0:10 + i])
    algo = rpt.Window(width=10, model='l1', min_size=2, jump=5).fit(data[:10 + i])
    print(data[10 + i] - expected_mean)
    result = algo.predict(pen=10)
    if len(result) > 1:
        print(result[0])
    print(result)
    # display
rpt.display(data, result)
plt.show()
```
